### Define simple helper functions
I wanted to implement those simple functions myself, just as an exercise. Please note that I'm aware that my code isn't considering edge-cases. The implementation of those stats functions was purely for fun, will use standard libraries in the future. 

```python
import matplotlib.pyplot as plt 
import seaborn as sns

sns.set(color_codes=True)

def single_attr(data_points, index):
    """ returns a sorted list of data_points with attr at index"""
    values = []
    for dp in data_points:
        values.append(dp[index])
    return sorted(values, key=float)

def min(values):
    if len(values) > 0:
        return values[0]
    
    return None

def first_quartile(values):
    # q1 is the mid-point between min and median
    first_half = values[:int(len(values)/2)]
    return median(first_half)

def mean(values):
    sum = 0
    for v in values:
        sum += v 
        
    return sum / len(values)

def median(values):
    # median is defined as the middle number if length is odd, 
    # or average of the middle two number of length is even 
    if len(values)%2 == 1:
        return values[int(len(values)/2)]
    
    return (values[int((len(values))/2)-1] + values[int(len(values)/2)]) / 2

def third_quartile(values):
    # q3 is the mid-point between max and median 
    second_half = values[int(len(values)/2):]
    return median(second_half)
    
def max(values):
    if len(values) > 0:
        return values[len(values)-1]
    
    return None

def print_plot_stats(data_points, index, plot_title, color):
    values = single_attr(data_points, index) 
    stats = {
                "min": min(values),
                "q1": first_quartile(values), 
                "mean": mean(values),
                "median": median(values), 
                "q3": third_quartile(values), 
                "max": max(values), 
            }
    print(stats)
    
    sns.distplot(values, color=color)
    plt.title(plot_title)
```

### Import iris data

```python
data_source = open('iris.data')

data_points = []

while True:
    line = data_source.readline().replace("\n", "")
    if len(line) == 0:
        break

    dp = line.split(",")
    
    # if this line isn't formatted properly, don't record it.
    if len(dp) != 5:
        continue
        
    dp[0] = float(dp[0])
    dp[1] = float(dp[1])
    dp[2] = float(dp[2])
    dp[3] = float(dp[3])
    
    data_points.append(dp)

data_source.close()
```

# Sepal length regardless of class (150 samples)

## Analysis:
The distribution is close to normal distribution and almost symmetrical. 

```python
print_plot_stats(data_points, 0, 'sepal length regardless of class', 'C1')
plt.show()
```

# Sepal width regardless of class (150 samples)

## Analysis:

```python
print_plot_stats(data_points, 1, 'sepal width regardless of class', 'C2')
plt.show()
```

# Petal length regardless of class (150 samples)

# Seprating data points into indviual classes

```python
setosa_points = list(filter(lambda dp: dp[4] == 'Iris-setosa', data_points))
versicoulour_points = list(filter(lambda dp: dp[4] == 'Iris-versicolor', data_points))
virginica_points = list(filter(lambda dp: dp[4] == 'Iris-virginica', data_points))
```

# Comparing sepal length across indivual classes (50, 50, 50 samples)

## Analysis:
- Setosa class is less spread than the other two classes
- Each distribution is symmetrical. 
- Clases cannot be easily distinguished sepal length

```python
print("Setosa summary: ", end="")
print_plot_stats(setosa_points, 0, 'sepal length', 'tab:blue')

print("Versicolour summary: ", end="")
print_plot_stats(versicoulour_points, 0, 'sepal length', 'tab:green')

print("Virginica summary: ", end="")
print_plot_stats(virginica_points, 0, 'sepal length', 'tab:red')

print("")

plt.xlabel("Setosa: blue, Versicolour: green, Virginica: red")
plt.show()
```
